```python
import matplotlib.pyplot as plt
import numpy as np

from hdbo.factory import function_factory
```

```python
ackley_func, search_space, global_min = function_factory(
    function_name="ackley",
    return_lp=False
)
```

```python
x_bounds = search_space.bounds[0]
y_bounds = search_space.bounds[1]
```

```python
x = np.linspace(x_bounds[0], x_bounds[1], 10000)
y = np.linspace(y_bounds[0], y_bounds[1], 10000)

X, Y = np.meshgrid(x, y)
```

```python
Z = ackley_func(X, Y)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

surf = ax.plot_surface(X, Y, Z, cmap="plasma")

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.title("Ackley Function")
plt.xlabel("x")
plt.ylabel("y")
```
